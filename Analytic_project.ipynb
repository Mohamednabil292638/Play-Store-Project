```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set up figure
plt.figure(figsize=(10,6))

# Enhanced distribution plot with a color palette
sns.histplot(data=df, x='Rating', bins=20, kde=True, color='dodgerblue', edgecolor='black')

# Titles and labels
plt.title('Distribution of App Ratings', fontsize=18, weight='bold')
plt.xlabel('Rating', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Annotating the mean
mean_rating = df['Rating'].mean()
plt.axvline(mean_rating, color='red', linestyle='--', linewidth=2)
plt.text(mean_rating + 0.1, 50, f'Mean: {mean_rating:.2f}', color='red', fontsize=12)

# Show plot
plt.grid(True, alpha=0.3)
plt.show()
```

```python
from matplotlib import cm

# Set up figure
plt.figure(figsize=(12,8))

# Bar chart with color gradient based on count
category_counts = df['Category'].value_counts()
colors = cm.Blues(category_counts / max(category_counts))  # Color mapping

sns.barplot(y=category_counts.index, x=category_counts.values, palette=colors)
plt.title('Number of Apps by Category', fontsize=18, weight='bold')
plt.xlabel('Number of Apps', fontsize=14)
plt.ylabel('Category', fontsize=14)

# Show grid and plot
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Pie chart for Content Rating distribution
content_rating_counts = df['Content Rating'].value_counts()

# Create custom color palette and explode one slice
colors = sns.color_palette('Set2')
explode = [0.1 if i == 0 else 0 for i in range(len(content_rating_counts))]  # Explode the first slice

plt.figure(figsize=(8,8))
plt.pie(content_rating_counts, labels=content_rating_counts.index, autopct='%1.1f%%', startangle=140, 
        colors=colors, explode=explode, shadow=True, textprops={'fontsize': 14})

# Title
plt.title('Content Rating Distribution', fontsize=18, weight='bold')

# Show plot
plt.show()
```

```python
# Replace 'Free' with 0 in the 'Installs' column
df['Installs'] = df['Installs'].replace('Free', '0')

# Remove any non-numeric characters like +, ,
df['Installs'] = df['Installs'].str.replace('[+,]', '', regex=True)

# Convert 'Installs' to numeric (int type)
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')

# Drop rows with missing or invalid 'Installs' values
df.dropna(subset=['Installs'], inplace=True)

# Convert to int for final clean version
df['Installs'] = df['Installs'].astype(int)

# Plot the boxplot
plt.figure(figsize=(12,8))
sns.boxplot(x='Category', y='Installs', data=df, palette='Spectral')
plt.title('Distribution of App Installs by Category', fontsize=18, weight='bold')
plt.xlabel('Category', fontsize=14)
plt.ylabel('Installs (Log Scale)', fontsize=14)
plt.xticks(rotation=90)
plt.yscale('log')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(12,8))

# Scatter plot with log scale and size mapped to app size
sns.scatterplot(x='Reviews', y='Rating', data=df, size='Size', hue='Rating', sizes=(20, 200), palette='coolwarm', alpha=0.7)

# Titles and labels
plt.title('Rating vs. Reviews (Size represents App Size)', fontsize=18, weight='bold')
plt.xlabel('Reviews (Log Scale)', fontsize=14)
plt.ylabel('Rating', fontsize=14)
plt.xscale('log')  # Log scale for reviews

# Show plot
plt.grid(True, alpha=0.3)
plt.legend(title="Rating", loc='upper left')
plt.show()
```
